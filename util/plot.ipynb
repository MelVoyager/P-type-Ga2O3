绘制一个关于原子个数N的收敛性图，测试使用哪个N更好

```python
# 由于执行状态已重置，需要重新定义数据并绘图

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['Noto Sans CJK JP'] 
mpl.rcParams['font.size'] = 10  
mpl.rcParams['axes.unicode_minus'] = False 

# 重新定义数据
n_values = [40, 120, 160, 20]
energy_values = [-0.31866464E+03, -0.95832004E+03, -0.12749441E+04, -0.15854181E+03]

# 计算平均能量
average_energy_values = [energy / n for energy, n in zip(energy_values, n_values)]

# 按 n 排序数据
sorted_indices = sorted(range(len(n_values)), key=lambda i: n_values[i])
n_values = [n_values[i] for i in sorted_indices]
average_energy_values = [average_energy_values[i] for i in sorted_indices]

# 重新绘图
plt.figure(figsize=(6, 4))
plt.plot(n_values, average_energy_values, marker='o', linestyle='-', color='b', label="每个原子平均能量")

# 添加标签和标题
plt.xlabel("原子数n")
plt.ylabel("每个原子平均能量")
plt.title("每个原子平均能量——原子数n")
plt.legend()
plt.grid()

# 显示图像
plt.show()
```

N_O的电荷转变能级

```python

# 数据准备
data = {
    r"N$_{O(I)}$": [4.03232, 3.542],
    r"N$_{O(II)}$": [2.88782, 2.599],
    r"N$_{O(III)}$": [3.83319, 3.363]
}

# 图形参数设置
x_positions = [1, 2, 3]  # 每列在x轴上的位置
colors = ['#2F5597', '#ED7D31']  # 不同来源的颜色
labels = ['我们的计算结果', r'文献$^1$中的计算结果']  # 图例标签
column_width = 0.8  # 每列占据的宽度

plt.figure(figsize=(8, 6))
ax = plt.gca()

# 绘制横线
for idx, (defect, values) in enumerate(data.items()):
    x_center = x_positions[idx]
    x_min = x_center - column_width/4
    x_max = x_center + column_width/4
    
    # 绘制两条能级线
    for i, value in enumerate(values):
        ax.hlines(y=value, 
                #  xmin=x_min + 0.05 + i*0.4,  # 横向偏移避免重叠
                #  xmax=x_max - 0.05 - (1-i)*0.4,
                xmin=x_min,
                xmax=x_max,
                 colors=colors[i], 
                 linewidth=3, 
                 label=labels[i] if idx == 0 else "")

# 坐标轴设置
ax.set_xticks(x_positions)
ax.set_xticklabels(data.keys(), fontsize=12)
plt.ylabel(r"$\varepsilon(0/-)$ eV", fontsize=12)
plt.ylim(2.5, 4.2)  # 设置Y轴范围

# 添加辅助网格
ax.grid(axis='y', linestyle='--', alpha=0.7)

# 图例设置
legend = ax.legend(loc='upper right', 
                  frameon=True,
                  edgecolor='black',
                  fontsize=10)
legend.get_frame().set_facecolor('white')

# 添加数据标签
for idx, (defect, values) in enumerate(data.items()):
    x_center = x_positions[idx]
    for i, value in enumerate(values):
        ax.text(x_center - 0.15 + 0.3*i, 
               value + 0.05, 
               f"{value:.2f} eV", 
               ha='center',
               color=colors[i],
               fontsize=10)

# 美化边框
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

绘制$P_O$电荷态转变能级

```python
# 数据准备（仅保留第一个来源的数据）
data = {
    r"P$_{O(I)}$": [5.29231],  # 只保留第一个值
    r"P$_{O(II)}$": [4.32178], 
    r"P$_{O(III)}$": [4.47017]
}

# 图形参数设置（简化为一个来源）
colors = ['#2F5597']  # 仅使用第一个颜色
labels = ['我们的计算结果']  # 仅保留第一个标签
column_width = 0.8

plt.figure(figsize=(8, 6))
ax = plt.gca()

# 绘制横线（简化循环）
for idx, (defect, values) in enumerate(data.items()):
    x_center = x_positions[idx]
    x_min = x_center - column_width/4
    x_max = x_center + column_width/4
    
    # 只绘制第一个来源的数据
    ax.hlines(y=values[0],  # 取列表中的第一个值
              xmin=x_min,
              xmax=x_max,
              colors=colors[0], 
              linewidth=3, 
              label=labels[0] if idx == 0 else "")

# 坐标轴设置（保持相同范围以维持比例）
ax.set_xticks(x_positions)
ax.set_xticklabels(data.keys(), fontsize=12)
plt.ylabel(r"$\varepsilon(0/-)$ eV", fontsize=12)
plt.ylim(3.5, 6)

# 辅助网格保持原样
ax.grid(axis='y', linestyle='--', alpha=0.7)

# 图例设置（仅显示一个图例）
legend = ax.legend(loc='upper right', 
                  frameon=True,
                  edgecolor='black',
                  fontsize=10)
legend.get_frame().set_facecolor('white')

# 数据标签（仅标注第一个来源）
for idx, (defect, values) in enumerate(data.items()):
    x_center = x_positions[idx]
    ax.text(x_center,  # 居中显示
           values[0] + 0.05, 
           f"{values[0]:.2f} eV", 
           ha='center',
           color=colors[0],
           fontsize=10)

# 美化边框（保持原样）
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
